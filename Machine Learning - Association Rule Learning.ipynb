```python
# ARL - Association Rule Learning
# Types -  Apriori ans Eclat

# Apriori
# support(M) = #user watchlist containing M / #user watchlists
# confidence(M1 -> M2) = #user watchlist containing M1 and M2 / #user watchlists containing M1
# lift(M1 -> M2) = confidence(M -> M2) / support(M2)

# Algorithm
# Step 1 - set a minimum number of support and confidence
# Step 2 - take all the subsets in transaction having higher support than minimum support
# Step 3 - take all the rules of these subsets having higher confidence than minimum confidence 
# Step 4 - sort all the rules by decreasing lift


# Eclat
# support(M) = #user watchlist containing M / #user watchlists

# Algorithm
# Step 1 - set a minimum support 
# step 2 - take all the subsets in transaction having higher support than minimum support
# Step 3 - sort all these subsets by decreasing support
```

# Apriori

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

dataset = pd.read_csv("/content/drive/MyDrive/Machine Learning Udemy/Market_Basket_Optimisation.csv",  header = None)
```

```python
!pip install apyori
```

```python
# Recreating the dataset in a certain format accepted by aprori model
# That format is list of transaction
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2 )           # 3*7/7501 = 0.003
```

```python
# Visualising The Results
results = list(rules)
results
```

```python
# using pandas for better visuals
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

```python
resultsinDataFrame
```

```python
## Displaying the results sorted by descending lifts
resultsinDataFrame.nlargest(n = 10, columns = 'Lift')
```

# Eclat

```python
# Eclat is a simplified version of apriori

!pip install apyori

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')

dataset = pd.read_csv("/content/drive/MyDrive/Machine Learning Udemy/Market_Basket_Optimisation.csv",  header = None)
```

```python
# Recreating the dataset in a certain format accepted by aprori model
# That format is list of transaction
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2 )           # 3*7/7501 = 0.003
```

```python
# Visualising The Results
results = list(rules)
results
```

```python
# using pandas for better visuals
def inspect(results):
    p1         = [tuple(result[2][0][0])[0] for result in results]
    p2         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    return list(zip(p1, p2, supports))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['product 1', 'product 2', 'Support'])
```

```python
## Displaying the results sorted by descending lifts
resultsinDataFrame.nlargest(n = 10, columns = 'Support')
```
